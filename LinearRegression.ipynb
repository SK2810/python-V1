```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("/content/insurance.csv")
```

```python
data.head() #it shows first 5 rows of the dataset
```

```python
data.info() #analyze the data
```

```python
data.describe() #gives statastical analysis
```

```python
#STEP 2: DATA PRE PROCESSING
#To convert catagorial variables into numerical variables
#Using ONE HOT encoding
data['smoker']=data['smoker'].apply({'yes':1,'no':0}.get)
```

```python
data.head()
```

```python
data['sex']=data['sex'].apply({'male':1,'female':0}.get)
```

```python
data.head()
```

```python
data['region']=data['region'].apply({'southwest':1,'southeast':2,'northwest':3,'northeast':4}.get)
```

```python
data.head() #all 'object data types are converted to int
```

```python
#STEP 3: division of data into Independent and Dependent columns
X=data[['age','sex','bmi','children','smoker','region']] #All dependent heads in X
y=data['charges']
```

```python
X.head()
```

```python
y.head()
```

```python
#Spliting the data into train and test ratios
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2) #this means test size is 20%
```

```python
#Create machine learning model 'Linear Regression'
from sklearn.linear_model import LinearRegression
linearmodel=LinearRegression()
linearmodel.fit(X_train,y_train) #training the model
```

```python
#predictions
linearmodel.predict(X_test)
```

```python
#Calculate accuracy of the model
linearmodel.score(X,y) #anything above 70 accuracy is decent
```

```python
#Predict with new data set
data_new={'age':67,'sex':1,'bmi':23,'children':2,'smoker':1,'region':3}
index=[1] #serial number
my_data=pd.DataFrame(data_new,index)
```

```python
my_data.head()
```

```python
#to check predicted charges for above data 
my_data_charges=linearmodel.predict(my_data)
print("The charges to be paid by the new customer ",my_data_charges)
```
